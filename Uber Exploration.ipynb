```python
import csv
```

```python
fn = 'uber-raw-data-apr14.csv'
```

```python
data = []
with open(fn) as f:
    for record in csv.DictReader(f):
        data.append(record)
    
```

```python
len(data)
```

```python
data[0]
```

```python
%pylab inline
```

```python
longitudes = []
for record in data:
    lon = record['Lon']
    longitudes.append(float(lon))
```

```python
longitudes[:5]
```

```python
hist(longitudes, range = (-74.05, -73.9), bins = 100)
```

```python
latitudes = []
for record in data:
    lat = record['Lat']
    latitudes.append(float(lat))
```

```python
latitudes[:10]
```

```python
hist(latitudes, range=(40.5, 41), bins = 100)
```

```python
figure(figsize=(20,20))
plot(longitudes, latitudes, '.', ms=1)
ylim(40.65, 40.8)
xlim(-74.05, -73.85)
```

```python
datetimes = []
for record in data:
    dt = record['Date/Time']
    datetimes.append(dt)
```

```python
datetimes[:10]
```

```python
dt = '4/1/2014 0:11:00'
```

```python
hours = []
for dt in datetimes:
    time = dt.split(' ')[1]
    hour = int(time.split(':')[0])
    hours.append(hour)
```

```python
hours[:10]
```

```python
hist(hours, bins=24)
grid()
xlabel('hour of the day')
ylabel('frequency of drop-off')
title('histogram of drop-offs by hour')
```

```python
import collections
```

```python
hours_counter = collections.Counter(hours)
```

```python
hours_counter
```

```python
hours_counter.most_common(5)
```

```python
collections.Counter(datetimes).most_common(20)
```

```python
dates = []
for dt in datetimes:
    date = dt.split(' ')[0]
    dates.append(date)
```

```python
by_date = collections.Counter(dates).most_common(31)
```

```python
counts = []
for r in by_date:
    counts.append(r[1])
counts
```

```python
bar(range(1,31),counts)
```

```python
labels = []
for val in by_date:
    date = val[0]
    dom = date.split('/')[1]
    labels.append(dom)
```

```python
bar(range(1,31),counts)
xticks(range(1,31), labels, rotation=75)
```
